```python
from rocketpy.Function import Function, PiecewiseFunction
import numpy as np
```

```python
tank_geometry = {(0, 1): lambda y: np.sqrt(1 ** 2 - (1 - y) ** 2), (1, 5): lambda y: 1, (5, 6): lambda y: np.sqrt(1 ** 2 - (y - 5) ** 2)}
```

```python
tank_rad = PiecewiseFunction(tank_geometry)
```

```python
tank_rad.plot1D()
```

```python
tank_area = np.pi * tank_rad ** 2
```

```python
tank_area.plot1D()
```

```python
tank_vol = tank_area.integralFunction()
```

```python
tank_vol.plot1D()
```

```python
tr = tank_vol.reverse()
```

```python
tr()
```
