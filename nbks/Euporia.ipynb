```python
# Sensitivity calculation trial function
def evaluateSensitivity(function, inputs, outputs, unit, depth):
    d = depth
    # Generate copy of inputs without deviations
    inputsCopy = inputs.copy()
    for key, value in inputs.items():
        inputsCopy[key] = value[0]
    # Evaluate sentivity of "function" to every "input" individually
    sensibilities = [[] for x in range(len(outputs))]
    for key, value in inputs.items():
        print('Evaluating sentivity of the desired function to ' + key + ' variation')
        print()
        dataSet = [[] for x in range(len(outputs))]
        for n in range(-1*depth, depth+1):
            n *= unit
            sample = inputsCopy.copy()
            x = value[0] + n*value[1]
            sample[key] = x
            ys = function(sample)
            for i in range(len(ys)):
                dataSet[i].append([x, ys[i]])
        for i in range(len(dataSet)):
            data = dataSet[i]
            # Spline Analysis
            dataFunction = Function(data, interpolation='spline')
            coeffs = dataFunction.__splineCoefficients__[:, d]
            zeroth = coeffs[0]
            first = coeffs[1]
            second = 2*coeffs[2]
            print('Analysis of ' + outputs[i] + ' x ' + key + ' - zeroth: ' + str(zeroth) + ' first: ' + str(first) + ' influence: ' + str(first*value[1])) 
            # Store data
            sensibilities[i].append([zeroth, first, second, value[1]])
            # Finite Difference Analysis
            zeroth = data[d][1]
            first = (data[d+1][1]-data[d-1][1])/(data[d+1][0]-data[d-1][0])
            second = 2*((data[d+1][1]-data[d][1])/(data[d+1][0]-data[d][0]) - (data[d-1][1]-data[d][1])/(data[d-1][0]-data[d][0]))/(data[d+1][0]-data[d-1][0])
            # print('FinDif analysis of ' + outputs[i] + ' x ' + key + ' - zeroth: ' + str(zeroth) + ' first: ' + str(first) + ' second: ' + str(second))
            # print()
    # Calculate general values
    minTotalDeviation = len(outputs)*[0]
    maxTotalDeviation = len(outputs)*[0]
    for i in range(len(outputs)):
        for j in sensibilities[i]:
            minTotalDeviation[i] += j[1]*j[3]
            maxTotalDeviation[i] += (j[1]*j[3])**2
    maxTotalDeviation = [i**0.5 for i in maxTotalDeviation]
    minTotalDeviation = [(i**2)**0.5 for i in minTotalDeviation]
    for i in range(len(outputs)):
        print(outputs[i])
        print("Minimum Total Deviation: " + str(minTotalDeviation[i]))
        print("Maximum Total Deviation: " + str(maxTotalDeviation[i]))
    return sensibilities
S = evaluateSensitivity(func, INPUTS, ['Apogee', 'xImpact', 'yImpact'], unit=0.5, depth=12)
```

```python
def func(inputs):
    Jiboia58 =  Motor(thrustSource='../data/jiboia/thrustCurve.csv',
                  burnOut=2.84,
                  reshapeThrustCurve=(inputs['burnOut'], inputs['impulse']),
                  interpolationMethod='spline',
                  nozzleRadius=inputs['nozzleRadius'],
                  throatRadius=inputs['throatRadius'],
                  grainNumber=5,
                  grainSeparation=inputs['grainSeparation'],
                  grainDensity=inputs['grainDensity'],
                  grainOuterRadius=inputs['grainOuterRadius'],
                  grainInitialInnerRadius=inputs['grainInitialInnerRadius'],
                  grainInitialHeight=inputs['grainInitialHeight'])
    m_aeroI = inputs['m_aero']
    m_recI = inputs['m_rec']
    m_seI = inputs['m_se']
    m_propI = inputs['m_prop']
    EuporiaI =  Rocket(motor=Jiboia58,
                       mass=m_aeroI+m_propI+m_recI+m_seI,
                       inertiaI=0.3462,
                       inertiaZ=0.00278,
                       radius=inputs['radius'],
                       distanceRocketNozzle=inputs['distanceRocketNozzle'],
                       distanceRocketPropellant=inputs['distanceRocketPropellant'],
                       offCenter=0,
                       powerOffDrag="../data/euporia/euporiaIDragOff.csv",
                       powerOnDrag="../data/euporia/euporiaIDragOn.csv",
                       drogueArea=False,
                       drogueCd=False,
                       drogueLag=1,
                       mainArea=False,
                       mainCd=False,
                       mainAlt=False)
    EuporiaI.powerOffDrag = inputs['powerOffDrag']*EuporiaI.powerOffDrag
    EuporiaI.powerOnDrag = inputs['powerOnDrag']*EuporiaI.powerOffDrag
    EuporiaI.addNose(length=inputs['noseLength'], kind='parabolic', distanceToCM=inputs['noseDistanceToCM'])
    EuporiaI.addTail(topRadius=inputs['tail1TopRadius'], bottomRadius=inputs['tail1BottomRadius'], length=inputs['tail1Length'], distanceToCM=inputs['tail1DistanceToCM'])
    EuporiaI.addTail(topRadius=inputs['tail2TopRadius'], bottomRadius=inputs['tail2BottomRadius'], length=inputs['tail2Length'], distanceToCM=inputs['tail2DistanceToCM'])
    EuporiaI.addFins(n=4, rootChord=inputs['finRootChord'], tipChord=inputs['finTipChord'], span=inputs['finSpan'], distanceToCM=inputs['finDistanceToCM'])
    EuporiaI.addTail(topRadius=inputs['tail3TopRadius'], bottomRadius=inputs['tail3BottomRadius'], length=inputs['tail3Length'], distanceToCM=inputs['tail3DistanceToCM'])
    FlightI = Flight(EuporiaI, E, inclination=inputs['inclination'], heading=inputs['heading'], flightPhases=-1, timeStep=[0.01, 0.1])
    return [FlightI.apogee, FlightI.xImpact, FlightI.yImpact]
```
