```python
from rocketpy import Function, LiquidMotor, MassFlowRateBasedTank, Fluid
```

```python
LiquidN2O = Fluid(name='Liquid Nitrous Oxide', density=855, quality=1)
VapourN2O = Fluid(name='Vapour Nitrous Oxide', density=101, quality=0)

N2O_tank = MassFlowRateBasedTank(name='Oxidizer Tank',
                                diameter=0.2,
                                height=1.6,
                                bottomCap='flat',
                                upperCap='flat',
                                initial_liquid_mass=42.45,
                                initial_gas_mass=0.95,
                                liquid_mass_flow_rate_in='../data/liquid_mass_flow_in.csv', #TO-DO: using this as a constant causes issues with Function class...
                                liquid_mass_flow_rate_out='../data/liquid_mass_flow_out.csv',
                                gas_mass_flow_rate_in='../data/gas_mass_flow_in.csv',
                                gas_mass_flow_rate_out='../data/gas_mass_flow_out.csv',
                                liquid=LiquidN2O,
                                gas=VapourN2O)
```

```python
N2O_tank.netMassFlowRate()

N2O_tank.liquid_mass_flow_rate_in.plot()
N2O_tank.liquid_mass_flow_rate_out.plot()
N2O_tank.liquid_net_mass_flow_rate.plot()
N2O_tank.net_mass_flow_rate()
```

```python
center_of_mass = Function(lambda t: N2O_tank.centerOfMass(t))
center_of_mass.plot(0,25)
```

```python
inertiaI = Function(lambda t: N2O_tank.inertiaTensor(t)[0])
inertiaZ = Function(lambda t: N2O_tank.inertiaTensor(t)[1])

inertiaI.setOutputs("Moment of Inertia I of Liquid (kg*m^2)")
inertiaI.plot(0,25)
```
